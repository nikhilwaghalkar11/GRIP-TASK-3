##### 1. Import the required libraires

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

##### 2. Importing Data

```python
df = pd.read_csv("SampleSuperstore.csv")
print("Data imported...üëç")
```

##### 3.Knowing data

```python
df.head() # seeing the head of data
```

```python
print(f"shape of the data is {df.shape}")
```

##### 4. EDA

```python
print("Basic info about data...")
df.info()
```

```python
print("Basic descriptive stats for numerical columns...")
df.describe().T
```

```python
print("counting classes in each categorical columns...")
for i in df.select_dtypes("O"):
    print("->",i,len(df[i].unique()))
```

##### 4.1 checking for nulls

```python
df.isnull().sum()
```

##### 4.3 Data Visualizations

```python
sns.pairplot(df) # pairplot to check relation in numerical columns
```

```python
print("heat map to check correlation in numerical columns...")
fig,axes = plt.subplots(1,1,figsize=(12,7))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

##### Sales and profit is highly correlated amoung all which quit obvious

##### 4.4 Univariate analysis 

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
sns.distplot(df["Profit"])
plt.subplot(1,2,2)
sns.distplot(df["Sales"])
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df["Discount"],bins=5)
plt.show()
```

##### 4.5 Multivariate analysis 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Sales",y="Profit",hue="Segment",data=df,size="Discount")
plt.title("sales vs profit ")
plt.show()
```

```python
sns.barplot(x="Region",y="Profit",data=df,estimator=np.sum)
plt.title("profit by Region")
plt.show()
```

```python
plt.figure(figsize=(15,6))
df.groupby(by=["Category","Sub-Category"])["Profit"].mean().plot(kind="bar")
plt.ylabel("profit")
plt.title("profit by categories and sub-categories")
plt.show()
```

```python
plt.figure(figsize=(15,6))
df.groupby(by=["Category","Sub-Category"])["Sales"].mean().plot(kind="bar")
plt.ylabel("profit")
plt.title("profit by categories and sub-categories")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(pd.crosstab(df["Region"],df["Quantity"]))
```

# Results of EDA

* #### The features Profit and Sales are highly related. 
* #### West and East region have more profit .
* #### The maximum quantity of product in demand was in range 2-4 in all regions.
* #### category technology has highest contribution in profit. 
* #### The Office supplies provides low Sales.
